```python
# compute gradient
def compute_gradient(f,x,y,w,b):
    '''
    this function computes gradient
    
    Args:
    f : f(w,b)
    x : train_x floor_size
    y : train(y) price
    w,b : model parameters
    
    Returns:
    dj_dw : gradient of cost J wrt w
    dj_db : gradient of cost J wrt b
    '''
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):
        dj_dw_i = (f[i] - y[i])*x[i]
        dj_db_i = f[i] - y[i]
        dj_dw += dj_dw_i
        dj_db += dj_db_i
    
    dj_dw = dj_dw/m
    dj_db = dj_db/m
    
    return dj_dw,dj_db
        
```

```python
def gradient_descent(x,y,w_in,b_in,alpha,num_iters):
    '''
    this function performs gradient descent to fit w and b by taking num_iters gradient steps with
    learning rate alpha
    
    Args:
    f : f(w,b)
    x : train_x floor(size)
    y : train_y price
    w_in,b_in : initial parameters
    alpha = learning rate 0.01
    num_inters : number of iterations
    
    Returns:
    w,b
    J_history : history of cost updates
    p_history : history of parameters updates
    
    '''
    w = w_in
    b = b_in
    J_history = []
    p_history = []
    
    for i in range(num_iters):
        f = train_func(w,b,x)
        dj_dw,dj_db = compute_gradient(f,x,y,w,b)
        cost_J = cost_func_J(f,y)
        
        w -= alpha*dj_dw
        b -= alpha*dj_db
        
        J_history.append(cost_J)
        p_history.append([w,b])
    
    return w,b,J_history,p_history
```
